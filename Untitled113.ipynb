```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.cluster import KMeans
import seaborn as sb
plt.style.use('seaborn-whitegrid')
```

```python
fulldata = pd.read_csv('https://raw.githubusercontent.com/dbeach24/USEnergy/master/cleaned_data/energy_all.csv')

fulldata = fulldata.rename(columns={'TETCB':'Tot_E', 'DFTCB':'Fuel_Oil', 'JFTCB':'Jet',
                                    'LGTCB':'LPG',    'MMTCB':'Gas', 'POTCB2':'Petrol_other',
                                    'NUETB':'Nuc',    'RETCB':'Renew', 'CLTCB':'Coal',
                                    'NNTCB':'NatG',   'PMTCB':'Petrol', 'WWTCB':'Bio',
                                    'EMTCB':'Ethanol', 'GETCB':'Geo',  'HYTCB':'Hydro',
                                    'SOTCB':'Solar',  'WYTCB':'Wind', 'ESTCB':'Elec'})
fulldata = fulldata.drop(columns=['Unnamed: 0','state.name','area','latitude','longitude'])

sources = ['Nuc','Fuel_Oil','Elec','Jet','import','export','LPG','Petrol_other','Gas',
           'NatG','Coal','Bio','Ethanol','Geo','Hydro', 'Solar','Wind']
```

```python
fulldata['Tot_E'] = 0
for i in sources:
    fulldata['Tot_E'] = fulldata['Tot_E'] + fulldata[i]
for i in sources:
    fulldata[i] = fulldata[i]/fulldata['Tot_E']
    
fulldata['Intensity'] = (fulldata['Tot_E']-(fulldata['export']*fulldata['Tot_E']))/(fulldata['gdp']*0.001)
```

```python
df = fulldata[['year','state','LPG']]
df = df.pivot(index='year',columns='state',values='LPG')
```

```python
X=df.T
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
labels=kmeans.labels_
X=df.T.rename_axis(None)

X['label']=labels

X=X.rename_axis(None)
```

```python
X[X.label==1].index
```

```python
#prints how many states in the kmean cluster were in region 1, region 2, etc.
```

```python
def Accuracy(actual):

    Acc=np.zeros(4)
    for i in range(4):
        for j in X[X.label==i].index:
            if j in actual:
                Acc[i]+=1
    return np.array(Acc)
```

```python
print(Accuracy(two))
print(Accuracy(one))
print(Accuracy(three))
print(Accuracy(four))
```
